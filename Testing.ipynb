```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-0.5, 0.5, 100)
y = np.linspace(-0.5, 0.5, 100)

a, b = np.meshgrid(x, y)
```

```python
theta = np.linspace(0, 2*np.pi, 1000)

radius = 1

a = radius*np.cos(theta)
b = radius*np.sin(theta)

figure, axes = plt.subplots(1)

axes.plot(a, b)
# axes.plot(a+0.2, b+0.2)
axes.set_aspect(1)

plt.title('Circle using Parametric Equation')
plt.show()
```

```python
from numpy import random

number_of_samples = 100000
rnd_thetas = np.array([random.random(number_of_samples) for i in range(2)])*2*np.pi
a1, b1 = np.cos(rnd_thetas[0]), np.sin(rnd_thetas[0])
a2, b2 = np.cos(rnd_thetas[1]), np.sin(rnd_thetas[1])
```

```python
yes, no = 0, 0

figure, axes = plt.subplots(1)

axes.plot(a, b)
# figure.set_size_inches(10.5, 10.5)

for i in range(number_of_samples):
    x1, y1, x2, y2  = a1[i], b1[i], a2[i], b2[i]
    _color =  "b" if (x1 - x2)**2 + (y1-y2) **2 > 3 else "r"
    if _color == 'b': 
        yes+= 1 
    else: 
        no+=1
    axes.plot([a1[i], a2[i]], [b1[i], b2[i]], color=_color)
#     plt.show()
axes.set_aspect(1)
plt.title('Circle using Parametric Equation')
# plt.show()

print("yes = {}, no = {}".format(yes, no))
print("ratio  = {}".format(yes/number_of_samples))
```

```python
number_of_samples = 100
rnd_thetas = random.random(number_of_samples)*2*np.pi
rnd_r = random.random(number_of_samples)

alpha = np.arccos(rnd_r)

alpa1, alpha2 = rnd_thetas + alpha, rnd_thetas - alpha

```

```python
rnd_thetas
```

```python
a1, b1 = np.cos(alpa1), np.sin(alpa1)
a2, b2 = np.cos(alpha2), np.sin(alpha2)
```

```python
yes, no = 0, 0

figure, (ax1, ax2) = plt.subplots(1, 2)

ax1.plot(a, b)
bars = ax2.bar(["yes" ,"no"], [0,0], color=["b", "r"])
# ax2.set_height(ax1.get_height())

# figure.set_size_inches(10.5, 10.5)

def animate(i):
    yes = bars[0].get_height()
    no = bars[1].get_height()
    x1, y1, x2, y2  = a1[i], b1[i], a2[i], b2[i]
    _color =  "b" if (x1 - x2)**2 + (y1-y2) **2 > 3 else "r"
    if _color == 'b': 
        yes+= 1 
    else: 
        no+=1
    ax1.plot([a1[i], a2[i]], [b1[i], b2[i]], color=_color)
    
    bars[0].set_height(yes)
    bars[1].set_height(no)
    max_height = max(bars[0].get_height(), bars[1].get_height()) + 5
    ax2.set_ylim((0, max_height))
    
#     plt.show()
ax1.set_aspect(1)
# ax2.set_aspect(1)

anim = animation.FuncAnimation(figure, animate,repeat=False,
                               frames=100, interval=100)

plt.show()

# print("yes = {}, no = {}".format(yes, no))
# print("ratio  = {}".format(yes/number_of_samples))
```
